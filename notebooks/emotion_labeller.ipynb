```python
from transformers import pipeline
```

```python
model = pipeline(model="seara/rubert-tiny2-cedr")
# model = pipeline(model="cointegrated/rubert-tiny2-cedr-emotion-detection")
# model = pipeline(model="seara/rubert-tiny2-ru-go-emotions")
label2russian = {
    "no_emotion": "нет эмоции",
    "joy": "радость",
    "sadness": "грусть",
    "surprise": "удивление",
    "fear": "страх",
    "anger": "злость",
}
```

```python
with open("GPT_FINAL.txt", "r") as f:
    array = f.readlines()
```

```python
processed_array = [
    x.strip().replace("User: ", "", 1).replace("AI: ", "", 1).replace("User:", "", 1).replace("AI:", "", 1) for x in array
]
```

```python
answer = []
i = 0
for text in processed_array:
    if text:
        if i == 0:
            answer.append("User: " + text + "\n")
            i += 1
        elif i == 1:
            answer.append("AI: " + text + "\n")
        answer.append("Emotion: " + label2russian[model(text)[0]["label"]] + "\n")
    else:
        i = 0
        answer.append("\n")
```

```python
answer
```

```python
with open("GPT_FINAL_EMOTIONS.txt", "w") as f:
    f.writelines(answer)
```
